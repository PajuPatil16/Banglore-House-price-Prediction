```python
def convertRange(x):

  temp = x.split('-')
  if len(temp) == 2:
    return (float(temp[0]) + float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
def removeOutliers(df):
  df_output = pd.DataFrame()
  for key, subdf in df.groupby('location'):
    m = np.mean(subdf.PricePerSQFT)

    st = np.std(subdf.PricePerSQFT)

    gen_df = subdf[((subdf.PricePerSQFT > (m-st)) & ((subdf.PricePerSQFT) <= (m+st)))]
    df_output = pd.concat([df_output,gen_df],ignore_index=True)
  return df_output
data = removeOutliers(data)
data.describe()
```

```python
def bhk_outlier_rem(df):
  exclude_indices = np.array([])
  for location, location_df in df.groupby('location'):
    bhk_sts = {}
    for bhk, bhk_df in location_df.groupby('bhk'):
      bhk_sts[bhk] =  {
          'mean' : np.mean(bhk_df.PricePerSQFT),
          'std' : np.std(bhk_df.PricePerSQFT),
          'count' : bhk_df.shape[0]
          }
    for bhk, bhk_df in location_df.groupby('bhk'):
      stats = bhk_sts.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.PricePerSQFT<(stats['mean'])].index.values)
  return df.drop(exclude_indices,axis='index')
```

```python
data = bhk_outlier_rem(data)
```

```python
x = data.drop(columns=['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print(x_train.shape,x_test.shape)
```

#Applying Linear Regression

```python
column_trans =  make_column_transformer((OneHotEncoder(sparse=False), ['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

# Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler, lasso)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_ls = pipe.predict(x_test)
r2_score(y_test,y_pred_ls)
```

# Applying Ridge

```python
ridge = Ridge()
pipe = make_pipeline(column_trans,scaler, ridge)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_rd = pipe.predict(x_test)
r2_score(y_test,y_pred_rd)
```

```python
print("No regulerization: ", r2_score(y_test, y_pred_lr))
print("Lasso: ", r2_score(y_test, y_pred_ls))
print("Ridge: ", r2_score(y_test, y_pred_rd))
```
